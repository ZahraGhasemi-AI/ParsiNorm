```python
def number_correction(word):
    for correct_form, wrong_forms in number_replaces.items():
        for wrong_character in wrong_forms:
            word = word.replace(wrong_character, correct_form)
    return word
```

```python
def select_templates(hour, minute, second):
    time_tempaltes = {
        0: f"{hour} Ÿà {minute} ÿØŸÇ€åŸÇŸá Ÿà {second} ÿ´ÿßŸÜ€åŸá",
        1: f"{words(hour)} Ÿà {words(minute)} ÿØŸÇ€åŸÇŸá Ÿà  {words(second)} ÿ´ÿßŸÜ€åŸá",
        2: f"{hour} Ÿà {minute}  Ÿà {second} ÿ´ÿßŸÜ€åŸá",
        3: f"{hour} Ÿà {minute}  ÿØŸÇ€åŸÇŸá Ÿà {second} ÿ´ÿßŸÜ€åŸá",
    }
    time_tempaltes_no_minute = {
        0: f"{words(hour)} Ÿà {words(second)} ÿ´ÿßŸÜ€åŸá",
        1: f"{hour} Ÿà {second} ÿ´ÿßŸÜ€åŸá",
        2: f"{hour} {second} ÿ´ÿßŸÜ€åŸá",
    }
    time_tempaltes_no_second = {
        0: f"{hour} Ÿà {minute} ÿØŸÇ€åŸÇŸá ",
        1: f"{words(hour)} Ÿà {words(minute)} ÿØŸÇ€åŸÇŸá",
        2: f"{hour} Ÿà {minute}",
        3: f"{hour} Ÿà {minute} ÿØŸÇ€åŸÇŸá ",
    }
    time_tempaltes_no_second_no_minute = {
        0: f"{hour}",
        1: f"{words(hour)}",
    }
    bamdad_tempaltes = {
        0: f"{minute} ÿØŸÇ€åŸÇŸá Ÿà {second} ÿ´ÿßŸÜ€åŸá ÿ®ÿßŸÖÿØÿßÿØ",
        1: f"{words(minute)} ÿØŸÇ€åŸÇŸá Ÿà  {words(second)} ÿ´ÿßŸÜ€åŸá ÿ®ÿßŸÖÿØÿßÿØ",
        2: f"{minute} Ÿà {second} ÿ´ÿßŸÜ€åŸá ÿ®ÿßŸÖÿØÿßÿØ",
        3: f"{minute} ÿØŸÇ€åŸÇŸá Ÿà {second} ÿ´ÿßŸÜ€åŸá ÿ®ÿßŸÖÿØÿßÿØ",
    }
    bamdad_tempaltes_no_minute = {
        0: f"{second} ÿ´ÿßŸÜ€åŸá ÿ®ÿßŸÖÿØÿßÿØ",
        1: f"{words(second)} ÿ´ÿßŸÜ€åŸá ÿ®ÿßŸÖÿØÿßÿØ",
    }
    bamdad_tempaltes_no_second = {
        0: f"{minute} ÿØŸÇ€åŸÇŸá ÿ®ÿßŸÖÿØÿßÿØ",
        1: f"{words(minute)} ÿØŸÇ€åŸÇŸá ÿ®ÿßŸÖÿØÿßÿØ",
    }
    if hour != 0 and minute != 0 and second !=0: 
        template_no = random.randint(0,3)
        return time_tempaltes.get(template_no)
    if hour != 0 and minute == 0 and second !=0: 
        template_no = random.randint(0,2)
        return time_tempaltes_no_minute.get(template_no)
    if hour != 0 and minute != 0 and second ==0: 
        template_no = random.randint(0,3)
        return time_tempaltes_no_second.get(template_no)
    if hour != 0 and minute == 0 and second ==0: 
        template_no = random.randint(0,1)
        return time_tempaltes_no_second_no_minute.get(template_no)
    if hour == 0 and minute != 0 and second !=0: 
        template_no = random.randint(0,3)
        return bamdad_tempaltes.get(template_no)
    if hour == 0 and minute == 0 and second !=0: 
        template_no = random.randint(0,1)
        return bamdad_tempaltes_no_minute.get(template_no)
    if hour == 0 and minute != 0 and second ==0: 
        template_no = random.randint(0,1)
        return bamdad_tempaltes_no_second.get(template_no)
    if hour == 0 and minute == 0 and second ==0: 
        return "ÿ®ÿßŸÖÿØÿßÿØ"
```
